```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
titanic = sns.load_dataset('titanic')
```

```python
# Histogram: Age distribution
plt.figure(figsize=(8, 5))
sns.histplot(titanic['age'].dropna(), bins=20, kde=True)
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Box Plot: Fare distribution
plt.figure(figsize=(8, 5))
sns.boxplot(y=titanic['fare'])
plt.title('Fare Distribution')
plt.ylabel('Fare')
plt.show()
```

```python
# Bar Chart: Survival counts
plt.figure(figsize=(8, 5))
sns.countplot(x='survived', data=titanic)
plt.title('Survival Counts')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter Plot: Age vs. Fare by Survival
plt.figure(figsize=(8, 5))
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic)
plt.title('Age vs. Fare by Survival')
plt.show()
```

```python
# Pair Plot: Selected features
sns.pairplot(titanic[['age', 'fare', 'pclass', 'survived']].dropna(), hue='survived')
plt.show()
```

```python
# Heatmap: Correlations
plt.figure(figsize=(8, 5))
sns.heatmap(titanic[['age', 'fare', 'pclass']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='pclass', y='survived', data=titanic)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.text(0, 0.6, 'Higher survival in 1st class', fontsize=10, color='blue')
plt.show()
```

```python
# Misleading: Truncated y-axis
plt.figure(figsize=(8, 5))
sns.barplot(x='pclass', y='survived', data=titanic)
plt.ylim(0.3, 0.7)  # Misleading scale
plt.title('Misleading: Survival Rate by Class')
plt.show()

# Corrected
plt.figure(figsize=(8, 5))
sns.barplot(x='pclass', y='survived', data=titanic)
plt.ylim(0, 1)  # Full scale
plt.title('Corrected: Survival Rate by Class')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic)
plt.title('Age vs. Fare by Survival', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Fare', fontsize=12)
plt.legend(title='Survived')
plt.show()
```
